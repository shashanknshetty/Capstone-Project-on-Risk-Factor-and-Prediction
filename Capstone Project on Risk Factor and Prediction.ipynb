```python
Underweight: BMI less than 18.5
Normal Weight: BMI 18.5 to 24.9
Overweight: BMI 25 to 29.9
Obese: BMI 30 or greater
```

```python
bmi_range = []
for i in df['bmi']:
    if i <18.5:
        bmi_range.append('Underweight')
    elif 18.5<=i<24.9:
        bmi_range.append('Normal Weight')    
    elif 25.0<=i<29.9:
        bmi_range.append('Over weight')
    else:
        bmi_range.append('High BMI')
```

```python
df['BMI_Range'] = bmi_range
```

By seeing the above results we get to know that RandomForest performing best compared to others , but it is Overfitted we are going to use Randomized Search and Grid Serach to overcome overfitting.

```python
#Randomized Search CV

# Number of trees in random forest
n_estimators = list(range(100,1300,100))
# Number of features to consider at every split
max_features = ['auto', 'sqrt']
# Maximum number of levels in tree
max_depth = [int(x) for x in np.linspace(5, 30, num = 6)]
# Minimum number of samples required to split a node
min_samples_split = [2, 5, 10, 15, 100]
# Minimum number of samples required at each leaf node
min_samples_leaf = [1, 2, 5, 10]
```

```python
# Create the random grid

random_grid = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf}
```

```python
from sklearn.model_selection import RandomizedSearchCV

rf_regressor=RandomForestRegressor()
rf_model=RandomizedSearchCV(estimator=rf_regressor,param_distributions=random_grid,
                            cv=3,random_state=0)
rf_model.fit(X_train_scaled,y_train)
```
